```python
lips_df = pd.read_excel(lips_path)
packstuecke_df = pd.read_excel(packstuecke_path)
tbl_mva_df = pd.read_excel(tbl_mva_path)
traffic_tracking_df = pd.read_excel(traffic_tracking_path)
vbak_df = pd.read_excel(vbak_path)
vttp_df = pd.read_excel(vttp_path)
```

```python
print("LIPS columns:", lips_df.columns)
print("Packstuecke columns:", packstuecke_df.columns)
print("TBL_MVA columns:", tbl_mva_df.columns)
print("Traffic_Tracking columns:", traffic_tracking_df.columns)
print("VBAK columns:", vbak_df.columns)
print("VTTP columns:", vttp_df.columns)
```

```python

merged_df = pd.merge(tbl_mva_df, packstuecke_df, left_on='ID', right_on='TBL_MVA_ID')
merged_df = pd.merge(merged_df, traffic_tracking_df, left_on='TBL_MVA_ID', right_on='TBL_MVA_ID')
```

```python
merged_df.head(10)
```

```python
print("merged_df columns:" ,merged_df.columns)
```

```python
# Dropping duplicates
merged_df.drop_duplicates(inplace=True)

# Handle missing values

merged_df.dropna(subset=['EINGANGSDATUM_UHRZEIT', 'BEREITGESTELLT_DATUM_UHRZEIT'], inplace=True)
merged_df.fillna(method='ffill', inplace=True)  # Forward fill to handle other missing values

# Ensuring correct data types
merged_df['EINGANGSDATUM_UHRZEIT'] = pd.to_datetime(merged_df['EINGANGSDATUM_UHRZEIT'])
merged_df['TA_DATUM_UHRZEIT'] = pd.to_datetime(merged_df['TA_DATUM_UHRZEIT'])
merged_df['TA_DATUM_UHRZEIT'] = pd.to_datetime(merged_df['BEREITGESTELLT_DATUM_UHRZEIT'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics
print(merged_df.describe())

# Distribution of processing times
merged_df['Processing_Time'] = (merged_df['TA_DATUM_UHRZEIT'] - merged_df['EINGANGSDATUM_UHRZEIT']).dt.total_seconds() / 3600
plt.hist(merged_df['Processing_Time'], bins=50)
plt.title('Distribution of Processing Times')
plt.xlabel('Processing Time (hours)')
plt.ylabel('Frequency')
plt.show()

# Correlation heatmap
numeric_cols = merged_df.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(12, 8))
sns.heatmap(merged_df[numeric_cols].corr(), annot=True, fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Datetime Features
merged_df['EINGANG_YEAR'] = merged_df['EINGANGSDATUM_UHRZEIT'].dt.year
merged_df['EINGANG_MONTH'] = merged_df['EINGANGSDATUM_UHRZEIT'].dt.month
merged_df['EINGANG_DAY'] = merged_df['EINGANGSDATUM_UHRZEIT'].dt.day
merged_df['EINGANG_HOUR'] = merged_df['EINGANGSDATUM_UHRZEIT'].dt.hour
merged_df['EINGANG_DAYOFWEEK'] = merged_df['EINGANGSDATUM_UHRZEIT'].dt.dayofweek

# Order Characteristics
merged_df['VOLUME'] = merged_df['LAENGE_IN_CM'] * merged_df['BREITE_IN_CM'] * merged_df['HOEHE_IN_CM']
merged_df['WEIGHT'] = merged_df['GEWICHT_IN_KG']

# Operational Indicators
merged_df['SONDERFAHRT'] = merged_df['SONDERFAHRT'].fillna(0).astype(int)
# merged_df['GEFAHRGUT'] = merged_df['GEFAHRGUT'].fillna(0).astype(int)

# Check the final DataFrame
print(merged_df.head())
```
